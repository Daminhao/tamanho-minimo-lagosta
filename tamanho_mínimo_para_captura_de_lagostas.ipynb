Lesgislação para o tamanho mínimo de captura de lagostas:
[link da Portaria](https://www.in.gov.br/en/web/dou/-/portaria-sap/mapa-n-221-de-8-de-junho-de-2021-324768133)

lagostalesg.png

```python
def verifica_lagosta( especie, 
                      comprimento_da_cauda,
                      comprimento_do_cefalo,
                     
                      comprimento_da_cauda_min_vermelha = 14,
                      comprimento_do_cefalo_min_vermelha = 8,
                     
                      comprimento_da_cauda_min_verde = 11,
                      comprimento_do_cefalo_min_verde = 6.5):
  
#Se for vermelha (Panulirus argus)
    if especie == 'vermelha':
      if comprimento_da_cauda >= comprimento_da_cauda_min_vermelha and comprimento_do_cefalo >= comprimento_do_cefalo_min_vermelha:
        print('Lagosta',especie,'com tamanho permitido')
      else:
        print('Lagosta',especie,'abaixo do tamanho minimo')

# se for verde  (Panulirus laevicauda) 
    if especie == 'verde':
      if comprimento_da_cauda >= comprimento_da_cauda_min_verde and comprimento_do_cefalo >= comprimento_do_cefalo_min_verde:
        print('Lagosta',especie,'com tamanho permitido')
      else:
        print('Lagosta',especie,'abaixo do tamanho minimo')
```

```python
#Pode mudar os valores das variaveis 
especie = 'vermelha' # verde ou vermelha

comprimento_da_cauda = 14
comprimento_do_cefalo = 8

verifica_lagosta(especie, comprimento_da_cauda, comprimento_do_cefalo)
```
